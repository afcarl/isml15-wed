# Popcorn challenge - logistic regression classifier

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('data/popcorn-features.csv')
test = pd.read_csv('data/popcorn-features-test.csv')
train.head()
```

```python
T = np.array(train["sentiment"])
X = np.array(train.iloc[:,2:])
N = X.shape[0]
perm = np.random.permutation(N)
d = 0.8

X_train = X[perm[:np.floor(d*N)]]
T_train = T[perm[:np.floor(d*N)]]

X_valid = X[perm[np.ceil(d*N):]]
T_valid = T[perm[np.ceil(d*N):]]

X_test = np.array(test.iloc[:,1:])
```

```python
def balanced_accuracy(predictions,truth):
    ''' Returns the balanced accuracy, (TP/P + TN/N)/2. '''
    pos = truth == 1
    neg = truth == 0
    true_pos = float(np.sum(predictions[pos]==1))
    true_neg = float(np.sum(predictions[neg]==0))

    return 0.5*(true_pos/np.sum(pos) + true_neg/np.sum(neg))
```

```python
acc = list()
for lam in [1,10,100]:
    classifier = linear_model.LogisticRegression(C=1./lam)
    classifier.fit(X_train,T_train)
    pred_valid = classifier.predict(X_valid)
    acc.append(balanced_accuracy(pred_valid,T_valid))
```

```python
plt.plot(acc)
plt.show()
```

```python
pred_test = classifier.predict(X_test)
output = pd.DataFrame(data={"id":test["id"],"sentiment":pred_test})
output.to_csv('predictions.csv',index=False,quoting=3)
```
